```python
resort_hotel=df[df['hotel']=="Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=="City Hotel"]
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
resort_hotel
```

```python
# Grouping and calculating mean
resort_hotel = resort_hotel.groupby("reservation_status_date")[['adr']].mean()

# Accessing separately
reservation_dates = resort_hotel.index  # Accessing the index, which contains the reservation dates
adr_means = resort_hotel['adr']  # Accessing the 'adr' column, which contains the mean values

# Displaying separately
print("Reservation Dates:")
print(reservation_dates)

print("\nADR Means:")
print(adr_means)
```

```python
plt.figure(figsize=(20,8))
plt.title("Averge Dily Rate in city and Resort Hotel",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel['adr'],label="City Hotel")
plt.legend(fontsize=20)
plt.show()

```

From the plotted data showing the average daily rate (ADR) for both the resort hotel and the city hotel over time, we can observe several potential insights:

Seasonality: There may be visible patterns in the ADR over time, indicating seasonal variations in hotel pricing. For example, higher ADRs during peak tourist seasons and lower ADRs during off-peak periods suggest that hotels adjust their pricing strategies based on demand fluctuations.

Price Differences: The comparison between the ADR of the resort hotel and the city hotel allows us to observe any consistent price differences between the two types of accommodations. A higher ADR for one type of hotel compared to the other may suggest differences in amenities, location, or target market.

Trends: Trends in ADR over time can provide insights into the overall performance and market conditions of the hotels. Increasing ADRs may indicate a healthy demand for hotel rooms or successful pricing strategies, while decreasing ADRs might signal challenges in attracting guests or competing with other accommodations.

Potential Events or Influences: Sudden spikes or drops in ADRs could be indicative of external factors such as special events, economic changes, or shifts in tourist preferences. These observations can help hotel managers understand the impact of such events on their pricing and revenue management strategies.

Opportunities for Improvement: Analyzing ADR trends can help identify periods of underperformance or missed revenue opportunities. Hotels can use this information to adjust their pricing strategies, offer promotions, or enhance their services to better meet guest expectations and maximize revenue.

Overall, visualizing the ADR data for both the resort hotel and the city hotel provides valuable insights into pricing dynamics, market trends, and opportunities for revenue optimization within the hospitality industry.

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_df = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='not cancelled')
plt.legend()
plt.show()
```
